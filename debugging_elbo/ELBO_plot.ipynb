```python
import os
import pickle
import copy

import pandas as pd
import seaborn as sns
import numpy
import torch

from sklearn.decomposition import FastICA

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib.lines import Line2D

import glob
import warnings
warnings.filterwarnings('ignore')
sns.set(style="whitegrid")

```

```python


base_dir='/Users/phumon/Documents/Research/nlu-test-sets/params_mvirt_sync'
file_name=f'lr-0.0001-steps-$5500-alpha-lognormal-identity-dim*_theta-normal-identity_nosubsample_1.00_0.40_particles8/train_elbo_losses.p'
exp_dir=os.path.join(base_dir, file_name)

print(exp_dir)
```

```python
sorted(list(glob.iglob(exp_dir)))
```

```python
dim_list=[]
final_elbo_list=[]
elbo_list=[]
min_elbo_list=[]
for filepath in glob.iglob(exp_dir):
    dim = int(filepath.split('/')[-2].split('_')[0].rsplit('-',1)[1][3:])
    dim_list.append(dim)
    
    with open(filepath, 'rb') as f:
        elbos =  pickle.load(f)
    elbo_list.append(elbos)
    min_elbo_list.append(min(elbos))
    final_elbo_list.append(elbos[-1])
    
```

```python
plot_data = list(zip(dim_list, final_elbo_list, min_elbo_list, elbo_list))
plot_data
```

```python
plot_data = list(zip(dim_list, final_elbo_list, min_elbo_list, elbo_list))
#plot_data.append((1, 1436.64 ))
plot_data.sort(key=lambda x: x[0])
dim, elbo_data, min_elbo_data,elbo_list = list(zip(*plot_data))
```
